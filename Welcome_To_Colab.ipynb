Task 3: Aggregating Quarterly Data
In our project, "Analyzing the UK's Economic Trends," this task involves aggregating monthly GDP data into quarterly averages. We perform this task to create a more manageable and meaningful time series representation of the data. By grouping data into quarters, we gain a clearer overview of economic trends and make it easier to spot patterns and fluctuations. This aggregated quarterly data is pivotal for our project as it simplifies analysis and helps us uncover deeper insights into the UK's economic performance.

```
print(data)
#Convert the 'Time Period' column in the DataFrame 'data' to datetime format using 'pd.to_datetime' with the specified date format ('/%m/%Y').
data['Time Period'] = pd.to_datetime(data['Time Period'], format='/%m/%Y')
data.set_index('Time Period', inplace=True)
```

```
print(data)
#Create a new DataFrame 'quarterly_data' by resampling the 'data' DataFrame using 'resample('Q').mean()'.
quarterly_data = data.resample('Q').mean()
print(quarterly_data)
```

Task 4: Analyzing Recession Periods
In our project, "Analyzing the UK's Economic Trends," this task involves identifying and visualizing recession periods. By tracking GDP growth over time and pinpointing when it falls below zero for two consecutive quarters, we can detect recessionary periods. Visualizing these recession periods is essential as it helps us understand how economic downturns affect the UK. This analysis is crucial for our project, allowing us to provide valuable insights into the economic challenges the UK has faced and how it has recovered over time.

```
# --- WRITE YOUR CODE FOR TASK 4 ---
import numpy as np
#Identify recession periods in the data using boolean conditions and store them in the 'Recession' column.
recession_condition = quarterly_data['GDP Growth']<0
quarterly_data['Recession'] = recession_condition

#Fill missing values in the 'Recession' column with 'False' using 'quarterly_data['Recession'].fillna(False, inplace=True)'.
quarterly_data['Recession'].fillna(False, inplace=True)
print(quarterly_data)

plt.figure(figsize=(10,6))

#Prepare data for plotting by converting the index and 'GDP Growth' values into NumPy arrays
index_array=quarterly_data.index.values
GDP_Growth_array=quarterly_data['GDP Growth'].values


#Use 'plt.plot()' to plot the 'GDP Growth' values with green color, labeling it as 'GDP Growth' and setting the line width to 2.
plt.plot(index_array,GDP_Growth_array, color='green', label='GDP Growth', linewidth=2 )

#Isolate recession data by filtering 'quarterly_data' for rows where 'Recession' is 'True'.
recession_data = quarterly_data[quarterly_data['Recession']==True]

#Extract the index and 'GDP Growth' values for the recession periods and plot them in red with the label 'Recession' and a line width of 2.
recession_index = recession_data.index.values
recession_GDP_Growth=recession_data['GDP Growth'].values
plt.plot(recession_index,recession_GDP_Growth, color='red', label='Recession', linewidth=2 )

#Set the title with 'plt.title('GDP Growth and Recession over Time (Quarterly Data)')', label the x-axis with 'plt.xlabel('Time Period')', label the y-axis with 'plt.ylabel('GDP Growth')', and add a legend using 'plt.legend()'.
plt.title('GDP Growth and Recession over Time (Quarterly Data)')
plt.xlabel('Time period')
plt.ylabel('GDP Growth')
plt.legend()
plt.show()
```
